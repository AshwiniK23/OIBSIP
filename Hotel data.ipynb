```python
def get_whiskers(col):
    q1 = np.quantile(col, 0.25)
    q3 = np.quantile(col, 0.75)            
    IQR = q3-q1
    lw = q1 - 1.5*IQR                          
    uw = q3 + 1.5*IQR                          
    return lw, uw
```

```python
def return_outliers_index(df,col):
    index_list = []
    lw, uw = get_whiskers(df[col])
    for m in df[col]:
        if m > uw:
            index = list(df[df[col] == m].index)
            index_list.extend(index)
        elif m < lw:
            index = list(df[df[col] == m].index)
            index_list.extend(index)
        else:
            pass
    return index_list              
```
